```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
page = requests.get('https://id.wikipedia.org/wiki/Daftar_bank_di_Indonesia')
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
soup
```

```python
tabel = soup.findAll('table',{'class':'wikitable'})[0]
baris = tabel.findAll('tr')
```

```python
for row in baris:
    for cell in row.findAll(['td','th']):
        cell.get_text()
```

```python
tampildata = []
for row in baris:
    info =[]
    for cell in row.findAll(['td','th']):
        info.append(cell.get_text().strip())
    tampildata.append(info)
```

```python
df = pd.DataFrame(tampildata)
display(df)
```

```python
df = pd.DataFrame(tampildata)
new_header = df.iloc[0]
df = df[1:]
df.columns = new_header
df.reset_index(drop = True, inplace = True)
display(df)
```

```python
df.to_csv('databank.csv')
```
